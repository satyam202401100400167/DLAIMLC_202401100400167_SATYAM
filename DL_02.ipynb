```python
x = [75, 65, 80] # Input features
w = [4, 3, 5]   # Weights
b = -1000       # Bias
eta = 0.01      # Learning rate
y = 1           # True label
```

```python
z = sum([a * b for a, b in zip(x, w)]) # Calculate weighted sum
z = z + b # Add bias
print(z) # Print intermediate result
```

```python
def step(z): # Define a step activation function
    return 1 if z >= 0 else 0
```

```python
y_pred = step(z) # Get prediction using the step function
print("Prediction (before learning):", y_pred)
```

```python
error = y - y_pred # Calculate prediction error
print("Error:", error)
```

```python
if error != 0: # Only update if there's an error
    for i in range(len(w)): # Update each weight
        w[i] = w[i] + eta * error * x[i]

    b = b + eta * error # Update the bias
```

```python
z = sum([a * b for a, b in zip(x, w)]) # Recalculate weighted sum with updated weights
z = z + b # Add updated bias
print(z) # Print new intermediate result
```
